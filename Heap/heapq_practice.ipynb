```python
# heapify: O(N) time
# heappop, heappush, heapreplace: log(N) in worst case
# Find min : O(1)
```

```python
from heapq import heappush, heappop, heapify, heapreplace, nsmallest
```

```python
a = []
heappush(a,29)
a
```

```python
heappush(a,14)
a
```

```python
heappush(a,7)
a
```

```python
heappush(a,1)
a
```

```python
# Heap sort in increasing order: Nlog(N)
a = [12, 32, 12, 4, 18, 23, 1, 29, 1]
heapify(a)
increasing_a = []

while a: 
    increasing_a.append(heappop(a))
increasing_a
```

```python
# Heap sort in descreasing order Nlog(N)
a = [12, 32, 12, 4, 18, 23, 1, 29, 1]
neg_a = [-x for x in a]
heapify(neg_a)
decreasing_a = []

while neg_a: 
    y = heappop(neg_a)
    decreasing_a.append(-y)
decreasing_a
```

```python
a = [12, 32, 12, 4, 18, 23, 1, 29, 1]
heapify(a)

nsmallest(4,a)
```

```python
# heapreplace : replace the min of heap with the new item

heap_n = [3,5,7,8,6,10,8]
heapreplace(heap_n,78)
heap_n
```
