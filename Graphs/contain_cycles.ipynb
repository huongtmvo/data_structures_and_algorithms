## Check if a graph contains cycles

```python
'''
For directed graphs, 

WHITE: never visited, 
GREY: visiting,
BLACK: done visited
'''

WHITE, GREY, BLACK = 0, 1, 2

class Solution:
    def explore(self, adj, v):
        self.visited[v] = GREY
        for u in adj[v]:
            if self.visited[u] == GREY: 
                return True
            if self.visited[u] == WHITE:
                if self.explore(adj, u):
                    return True   
        self.visited[v] = BLACK  
        return False        
            
    def acyclic(self, adj):
        self.visited = [WHITE]*len(adj)
        for v in range(len(adj)):
            if self.visited[v] == WHITE:
                if self.explore(adj, v): 
                    return 1
        return 0 
```

```python
adj = {0:[1],
       1:[2],
       2:[3],
       3:[]}
Solution().acyclic(adj)
```

```python
adj = {0:[1],
       1:[2],
       2:[3],
       3:[0]}
Solution().acyclic(adj)
```
