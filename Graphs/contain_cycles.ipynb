## Check if a graph contains cycles

An undirected graph with four vertices and five edges:

4 5 

2 1 

4 3 

1 4 

2 4 

3 2

```python
adj = {1:[2,4],
       2:[1,3,4],
       3:[2,4],
       4:[1,2,3]}
```

```python
'''
WHITE: not yet visited
GREY: visiting
BLACK: finished visiting
'''

WHITE, GREY, BLACK = 0,1,2
def explore(adj, v):
    visited[v] = GREY
    for u in adj[v]:
        if visited[u] == GREY:
            return True
        if visited[u] == WHITE:
            explore(adj,v)
    visited[v] = BLACK
    return False
```

```python
n = len(adj)
visited = [WHITE] *(n+1)
explore(adj, 1)
```
